```python
import cv2
import numpy as np
from PIL import Image

im_path = "image1.png"

im_array = cv2.imread(im_path)

# Convert to HSV color space for better color thresholding
hsv = cv2.cvtColor(im_array, cv2.COLOR_BGR2HSV)

# Define lower and upper red color ranges (in HSV color space)
lower_red = np.array([0, 50, 50])
upper_red = np.array([10, 255, 255])
lower_red2 = np.array([170, 50, 50])
upper_red2 = np.array([180, 255, 255])

# Create a mask of the red pixels within the circle
mask = cv2.inRange(hsv, lower_red, upper_red) + cv2.inRange(hsv, lower_red2, upper_red2)
display(Image.fromarray(cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)))

# Apply the mask to the original image
im_array[mask==255] = [0, 0, 0]

display(Image.fromarray(cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)))
```
