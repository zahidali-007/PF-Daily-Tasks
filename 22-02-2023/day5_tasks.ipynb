### Problem # 01 Find out highest and lost salary for Data Analyst with company_size

#### Import Libraries and dataset

```python
# importing libraries
import pandas as pd
```

```python
# load our dataset
df = pd.read_csv('ds_salaries.csv')
df.head()
```

```python
# To see columns names
df.columns
```

```python
# First we need to extract Data Analyst role from job_title column
data_analyst = df[df["job_title"] == 'Data Analyst']
data_analyst.head()
```

```python
# Here we store the salary and company_size, and job_title in one veriable data_analyst_salary
data_analyst_salary = data_analyst[['salary','company_size','job_title']]

# Here find the highest salary
highest_salary = data_analyst_salary.max()
# Here find the lowest salary
lowest_salary = data_analyst_salary.min()
print("Highest salary and company size:",highest_salary)
print("Lowest salary and company size:",lowest_salary)
type(highest_salary)
```

```python
# Here we convert pur above data into a dataframe
data =pd.DataFrame([highest_salary,lowest_salary])
data.head()
```

```python
type(data)
```

### Problem # 02 Sort the region as per have more jobs data engineer

```python
# Here we extract data of Data Engineer using job_title column
data_engineer = df[df["job_title"] == 'Data Engineer']
data_engineer.head()
```

```python
# Here we store the company_location, and job_title in one veriable data_engineer
data_engineer_loc = data_engineer[['company_location','job_title']]

# Here we used sort_values method to sort the values by region 
region = data_engineer_loc.sort_values(by='company_location',ascending=False)
region
```

```python
# Here we create a dataframe of our region
region_df = pd.DataFrame(region)
type(region_df)
```

```python
# Concatination of problem # 01 solution and problem # 02 solution
result = pd.concat([data,region_df])
result
```

```python
# Renaming columns names
result = result.rename(columns={'salary':'salary_range','company_location':'region'})
result.head()
```
