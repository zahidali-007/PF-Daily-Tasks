```python
import json

# Load the VGG annotation file
with open('annotation/file.json', 'r') as f:
    vgg_data = json.load(f)

# Create a dictionary in COCO format
# coco_data = {
#     "info": {},
#     "licenses": [],
#     "images": [],
#     "annotations": [],
#     "categories": []
# }

coco_data = {
    "info": {
        "description": "My Dataset",
        "url": "",
        "version": "1.0",
        "year": 2023,
        "contributor": "",
        "date_created": "2023-03-21"
    },
    "images": [],
    "annotations": [],
    "categories": [
        {
            "id": 1,
            "name": "class1",
            "supercategory": ""
        },
        {
            "id": 2,
            "name": "class2",
            "supercategory": ""
        },
        {
            "id": 3,
            "name": "class3",
            "supercategory": ""
        }
    ]
}

# Extract image information from the VGG annotation file and add it to the "images" list
for image in vgg_data["images"]:
    coco_image = {
        "id": image["id"],
        "width": image["width"],
        "height": image["height"],
        "file_name": image["file_name"]
    }
    coco_data["images"].append(coco_image)

# Extract annotation information from the VGG annotation file and add it to the "annotations" list
for annotation in vgg_data["annotations"]:
    coco_annotation = {
        "id": annotation["id"],
        "image_id": annotation["image_id"],
        "category_id": annotation["category_id"],
        "bbox": annotation["bbox"],
        "iscrowd": annotation["iscrowd"]
    }
    coco_data["annotations"].append(coco_annotation)

# Create a dictionary for each category and add it to the "categories" list
for category in vgg_data["categories"]:
    coco_category = {
        "id": category["id"],
        "name": category["name"]
    }
    coco_data["categories"].append(coco_category)

# Save the COCO dictionary as a JSON file
with open('coco/file.json', 'w') as f:
    json.dump(coco_data, f)
```
